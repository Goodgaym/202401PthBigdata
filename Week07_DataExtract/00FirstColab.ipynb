```python
import gdown
filename = 'PtLibraryData.csv'
url = 'https://drive.google.com/uc?id=14EDAU0RhBrloiwS_ieDo7xrKb4w5wkmj'
gdown.download(url, filename, quiet=False)
```

```python
# csv 파일의 인코딩을 확인할 수 있는 detect() 메서드를 가지고 있는 chardet 패키지
import chardet

# csv 파일을 읽어 변수 d에 저장하고 d의 인코딩을 확인해본다.
with open('PtLibraryData.csv', mode='rb') as f:
  d = f.readline()
print(chardet.detect(d))

# csv 파일을 읽어와 출력해보기
with open('PtLibraryData.csv') as f:
  print(f.readline())
  print(f.readline())
```

```python
# 데이터 프레임을 사용할 수 있는 pandas 패키지
import pandas as pd

# csv파일을 읽어 데이터프레임 객체를 생성하는 read_csv() 메서드
df = pd.read_csv('PtLibraryData.csv', encoding='utf-8', low_memory=False)
```

```python
# to_csv()메서드를 이용해 데이터프레임 csv로 저장 
df.to_csv('ns_202303.csv')
```

```python
# to_csv로 저장한 데이터프레임 csv 파일 읽어오기
with open('ns_202303.csv') as f:
  for i in range(3) :
    print(f.readline(), end='')
```

```python
# 데이터프레임을 csv 저장한 파일 읽기
ns_df = pd.read_csv('ns_202303.csv', low_memory=False)
ns_df.head()
```

```python
# index_col 매개변수로 인덱스의 위치를 지정해주고 읽기
ns_df = pd.read_csv('ns_202303.csv', index_col=0, low_memory= False)
ns_df.head()
```
