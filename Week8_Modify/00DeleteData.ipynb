```python
#CSV 읽어오기
import pandas as pd

pt_df = pd.read_csv(filename_drop, low_memory=False)
pt_df.head()
```

```python
#데이터프레임의 일부만 가져오기 (loc 함수)
pt_df.loc[0:1, ['도서명','대출건수']]
```

```python
#columns를 이용하여 잘못된 열 가져오기
pt_df.columns != 'Unnamed: 13'
```

```python
#잘못된 열을 제외한 맞는 열 만으로 데이터프레임 재구성
selected_columns = pt_df.columns != 'Unnamed: 13'
pt_book = pt_df.loc[:, selected_columns]
pt_book.head()
```

```python
#또는 drop함수를 이용하여 데이터프레임의 특정 열/행 제거
pt_book = pt_df.drop('Unnamed: 13', axis=1)
pt_book.head()
```

```python
#drop 함수의 inplace 속성을 이용하면 대입없이 바로 적용
pt_book.drop('권', axis=1, inplace = True)
pt_book.head()
```

```python
#dropna 함수 : NaN 데이터가 있는 열/행을 제거
pt_book = pt_df.dropna(axis=1)
pt_book.head()
```

```python
#dropna 함수의 how 매개변수 = all 일경우 모든 요소가 NaN일 경우 제거 
pt_book = pt_df.dropna(axis=1, how='all')
pt_book.head()
```

```python
#drop 함수를 이용하여 특정 인덱스만 제거
pt_tmp = pt_book.drop([0, 1])
pt_tmp.head()
```

```python
#또는 인덱싱을 이용하여 제거
pt_tmp = pt_book[2:]
pt_tmp.head()
```

```python
#다른 csv파일 불러오기
pt_book = pd.read_csv(filename, low_memory=False)
pt_book.head()
```

```python
#인덱스에 비교 연산자를 이용해 특정 출판사만을 True로 가져옴
selected_rows = pt_book['출판사'] == '문학동네'
print(selected_rows)
```

```python
#선택한 특정 출판사만 저장
pt_tmp = pt_book[selected_rows]
pt_tmp.head()
```

```python
#조건을 넣어 분류도 가능
pt_tmp = pt_book[pt_book['대출건수'] > 300]
pt_tmp.head()
```

```python
#도서명, 저자, ISBN의 중복데이터의 합계 *dup_list는 중복 데이터의 세트
dup_list = ['도서명', '저자', 'ISBN']
sum(pt_book.duplicated(subset = dup_list))
```

```python
#groupby 함수 : 데이터를 특정 값을 기준으로 그룹화 하는 함수, 데이터들을 연산해서 묶을 수도 있다.
count_df = pt_book[['도서명', '저자', 'ISBN', '대출건수']]

'''
group_df = count_df.groupby(by=['도서명', '저자', 'ISBN'], dropna=False)
loan_count = group_df.sum()
'''

loan_count = count_df.groupby(by=['도서명', '저자', 'ISBN'], dropna=False).sum()
loan_count.head()
```

```python
#중복되는 데이터 값들을 제외한 책들만 묶기
dup_rows = pt_book.duplicated(subset = ['도서명', '저자', 'ISBN'])
unique_rows = ~dup_rows
pt_book3 = pt_book[unique_rows].copy()
pt_book3.head()
```

```python
#인덱스를 특정 항목으로 지정 (도서명, 저자, ISBN)
pt_book3.set_index(['도서명', '저자', 'ISBN'], inplace=True)
pt_book3.head()
```

```python
#update 함수 : 같은 인덱스에 해당하는 값들을 수정
#그룹화를 이용해 중복되는 책의 대출건수를 갱신
pt_book3.update(loan_count)
pt_book3.head()
```

```python
#reset_index 함수 : 인덱스를 다시 초기화
pt_book4 = pt_book3.reset_index()
pt_book4.head()
```

```python
# 수정전 데이터프레임의 대출건수 확인
sum(pt_book['대출건수']>300)
```

```python
# 수정하여 다시 만든 데이터프레임의 대출건수 확인
sum(pt_book4['대출건수']>300)
```
