```python
import gdown
import pandas as pd
import matplotlib.pyplot as plt

# 아래 두개의 값 입력 (로컬 컴퓨터 환경에서 실행은 url 비워둘 것)
filename = 'pt_book_fix.csv'  # 예제에서 사용할 filename 입력
url = ''    # 여기에 본인이 가진 링크의 난수 입력

if url != '' :
    gdown.download(f'https://drive.google.com/uc?id={url}', filename, quiet=False)
    dir = ''
else :
    dir = '../_csv/'  # 로컬 컴퓨터 환경에서 사용할 dir (예제는 _csv)

pt_book = pd.read_csv(f'{dir}{filename}', low_memory=False)
pt_book.head()
```

## 히스토그램

```python
plt.hist([0,3,4,7,8,9,9,13], bins=5)    # 5개 구간의 히스토그램
plt.show()
```

```python
plt.hist(pt_book['대출건수'])
plt.show()
```

큰값과 작은값의 차이가 너무 커서 위와 같이 나옴
Y축을 로그스케일로 간격을 늘려본다.

```python
plt.hist(pt_book['대출건수'])
plt.yscale('log')
plt.show()
```

x축도 세세하게 나누면 아래와 같음

```python
plt.hist(pt_book['대출건수'], bins=100)
plt.yscale('log')
plt.show()
```

```python
title_len = pt_book['도서명'].apply(len)
plt.hist(title_len, bins=100)
plt.xscale('log')   # x도 로그스케일 추가
plt.show()
```

## 상자 수염 그림
- 상자 수염 그림은 최솟값, 세 개의 사분위수, 최댓값 의 5개의 숫자를 사용해 데이터를 요약   
직사각형 (25%~ 75% 지점)   
각각 50%기준 1.5배 멀리 띄워진 거리 수직선 긋고 수직선 밖에서 최솟값과 최댓값까지 데이터 점으로 표시   

이상치를 확인할때 좋음

```python
plt.boxplot(pt_book[['대출건수', '도서권수']])  # vert=False(수평), whis=10 (수염길이 조정)
plt.yscale('log')   # yscale 추가
plt.show()
```
