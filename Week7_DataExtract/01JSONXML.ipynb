JSON : JavaScript Object Notation    
```
{
  "Key1" : "Value1",
  "Key2" : "Value2"
}
```


````python
d = {"name":"데이터 분석"}
print(d['name'])
print(type(d))
````

````python
import json     # json을 파이썬에서 사용하기 위해 패키지 import

d_str = json.dumps(d, ensure_ascii=False)   # 한글을 사용하기 위하여 ensure_ascii 매개변수 False로 지정
print(d_str)        # d_str 출력 (문자열 형식)
print(type(d_str))  # d_str의 타입 출력 (str 타입)
````

````python
d2 = json.loads(d_str)
print(d2['name'])   # d2의 'name' 출력
print(type(d2))     # d2의 타입 출력 (dict 타입)
````

````python
d3_str = '''
{
    "name": "데이터 분석",
    "author": ["홍길동", "김철수"],
    "year": 2002
}
'''
d3 = json.loads(d3_str)
print(d3['author'][1])
````

````python
d4_str = """
[
  {"name": "데이터 분석", "author": "홍길동", "year": 2022},
  {"name": "머신러닝+딥러닝", "author": "홍길동", "year": 2020}
]
"""
d4 = json.loads(d4_str)
print(d4[0]['name'])
````

````python
import pandas as pd
import io

#json_df = pd.read_json(d4_str)  # json -> DataFrame 
json_df = pd.read_json(io.StringIO(d4_str))    # StringIO로 문자열을 파일 객체로 변환하여 json 읽어옴
list_df = pd.DataFrame(d4)      # Python 리스트 -> DataFrame

print(json_df)
print(type(json_df))
print(list_df)
print(type(list_df))
````

````python
x_str = """
<book>
    <name>데이터 분석</name>
    <author>홍길동</author>
    <year>2022</year>
</book>
"""
````

````python
import xml.etree.ElementTree as et

book = et.fromstring(x_str)

print(type(book))
print(book.tag)
````

````python
book_childs = list(book)

print(book_childs)
````

````python
name = book.findtext('name')
author = book.findtext('author')
year = book.findtext('year')

print(name)
print(author)
print(year)
````

````python
x2_str = """
<books>
  <book>
      <name>데이터 분석</name>
      <author>홍길동</author>
      <year>2022</year>
  </book>
  <book>
      <name>머신러닝+딥러닝</name>
      <author>김철수</author>
      <year>2020</year>
  </book>
</books>
"""

books = et.fromstring(x2_str)
````

````python
for book in books.findall('book'): 
  name = book.findtext('name')
  author = book.findtext('author')
  year = book.findtext('year')
  
  print(name)
  print(author)
  print(year)
````
